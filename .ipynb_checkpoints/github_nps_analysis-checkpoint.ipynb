### Import environment variables

```python
with open('.env') as f:
    db_url = f.readline().split('=')[1].strip().replace('"','')
```

```python
# Create SQLAlchemy Engine to Query DB
engine = create_engine(db_url)
```

```python
# Function to simplify writing and querying database
def dwh_execute(query):
    with open('.env') as f:
        for line in f:
            line = line.split('=')[1].strip().replace('"','')
            dbname = line[-6:]
            user = line.split('://')[1][:7]
            password = line.split('://')[1][8:].split('@')[0]
            host = line.split('@')[1].split(':')[0]
            port = line.split('@')[1].split(':')[1].split('/')[0]
            conn = psycopg2.connect(dbname=dbname, user=user, password=password, host=host, port=port,connect_timeout = 0)
            cur = conn.cursor()
            cur.execute(query)
            conn.commit()
```

### Read Github survey results .csv file to pandas dataframe

```python
git_df = pd.read_csv('/Users/enelson/Downloads/nps-exercise.csv')
```

```python
measures = ['github_in_org','github_in_pade_org','github_personal_paid', 'github_age_in_days', 'answer-10', 'answer-9','answer-15']
for measure in measures:
    for i, row in git_df.iterrows():
        if git_df.at[i, measure] == 't':
            git_df.at[i, measure] = True
        elif git_df.at[i, measure] == 'f':
            git_df.at[i, measure] = False
        elif measure == 'github_age_in_days':
            git_df.at[i, measure] = git_df.at[i, measure].replace(',','')
        elif git_df.at[i, measure] == 'Yes':
            git_df.at[i, measure] = True
        elif git_df.at[i, measure] == 'No':
            git_df.at[i, measure] = False
        elif git_df.at[i, measure] == 'Yes.':
            git_df.at[i, measure] = True
        elif git_df.at[i, measure] == 'No thanks!':
            git_df.at[i, measure] = False
```

```python
#Rename columns to make compatible with SQL column naming conventions
for column in git_df.columns:
    git_df.rename(columns={column:column.lower().replace('-','_')}, inplace=True)
```

```python
#Remove long varchar fields (1,000+ Characters from dataframe)
git_df = git_df.drop(['answer_11','answer_45','answer_59'], axis=1)
```

## Github's Net Promoter Score, Average Recommender Score, & Margin of Error

### Remove null recommender scores

```python
query = f'''
SELECT
count(id) as total_responses,
avg(answer_10) as avg_recommender_score,
COUNT(CASE WHEN answer_10 >= 9 then id else null end) as promoter_count,
(COUNT(CASE WHEN answer_10 >= 9 then id else null end)::float/COUNT(id)::float) as promoter_pct,
COUNT(CASE WHEN answer_10 <= 6 then id else null end) as detractor_count,
(COUNT(CASE WHEN answer_10 <= 6 then id else null end)::float/COUNT(id)::float) as detractor_pct,
COUNT(CASE WHEN answer_10 > 6 and answer_10 < 9 then id else null end) as passive_count,
(COUNT(CASE WHEN answer_10 > 6 and answer_10 < 9 then id else null end)::float/COUNT(id)::float) as passive_pct,
(COUNT(CASE WHEN answer_10 > 6 and answer_10 < 9 then id else null end)::float/COUNT(id)::float) + (COUNT(CASE WHEN answer_10 <= 6 then id else null end)::float/COUNT(id)::float) + (COUNT(CASE WHEN answer_10 >= 9 then id else null end)::float/COUNT(id)::float) as count_pct_check,
(COUNT(CASE WHEN answer_10 >= 9 then id else null end)::float/COUNT(id)::float) - (COUNT(CASE WHEN answer_10 <= 6 then id else null end)::float/COUNT(id)::float) as nps
from scratch.git_nps
where answer_10 is not null;
'''

df = pd.read_sql(query, engine)
```

### Variance, Standard Deviation, and Margin of Error

```python
var = ((1 - df.iloc[0]['nps'])**2)*df.iloc[0]['promoter_pct']+((0-df.iloc[0]['nps'])**2)*df.iloc[0]['passive_pct']+((-1-df.iloc[0]['nps'])**2)*df.iloc[0]['detractor_pct']
print(f'''Github's NPS Variance is: {var}\nGithub's NPS Standard deviation is: {math.sqrt(var)}\nGithub's NPS MoE (Margin of Error) is: {(math.sqrt(var))/math.sqrt(df.iloc[0]['total_responses'])}''')
```
